## To destroy the importance of all the work I did previously, these following 4 lines of code achieve the same purpose for all the features in the dataset. Here, the purpose is to identify outliers in the dataset since I assume they would be the most interesting calls. 

```python
res_out = pd.DataFrame(data=None, columns=df.columns)
for column in df.columns:
    res_out = pd.concat([res_out, df[(df[column]> df[column].mean() + 2.5*df[column].std())]])
```

```python
res_out.call_id.value_counts()
```

We can also find the most normal results and find the calls which behave most 'normal' across all features

```python
res_norm = pd.DataFrame(data=None, columns=df.columns)
for column in df.columns:
    res_norm = pd.concat([res_norm, df[(df[column]<df[column].quantile(0.55)) & (df[column]> df[column].quantile(0.45))]])
```

```python
res_norm.call_id.value_counts()
```

```python
pd.concat([res_out, res_norm]).call_id.value_counts()
```

We also make a correlation map to see if any two features aren't very coorelated since otherwise it'll repeat some datasets in the above result. And after inspecting the graph, we can safely assume that these features aren't highly coorelated to produce bad results. 

```python
corr = df.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
f, ax = plt.subplots(figsize=(15, 15))

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
